# 카테고리 변수 처리

```python
# gender 컬럼만 원핫 인코딩 해보기
# pd.get_dummies : 원핫 인코딩 함수
pd.get_dummies(data, columns=['gender'], drop_first = True) 
# drop_first: 첫째 컬럼 줄이기
```

## 카테고리 변수 전체 원핫 인코딩

```python
# 카테고리 변수 리스트 만들기
col_list=[]

for i in data.columns:
    if data[i].dtype == 'O':
        col_list.append(i)
```

```python
col_list
```

```python
# 컬럼 별 unique 값 계산
for i in col_list:
    print(i, data[i].nunique())
```

```python
# Customer는 카테고리 변수가 아니므로 제외
col_list = col_list[1:]
```

```python
col_list
```

```python
# 카테고리 변수 전체 -> 원핫 인코딩 : 분류 이므로 다중공선성 문제를 걱정하지 않아도 된다.
data_new = pd.get_dummies(data, columns=col_list, drop_first = True)
```

```python
print(data_new.shape)
data_new.head()
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler
```

```python
data_new.info()
```

### 1) Min-Max Scaler

```python
minmax = MinMaxScaler()
```

```python
# 불필요한 변수 제거
data_new.drop('customerID', axis=1, inplace =True)
```

```python
# fit : 데이터 특성값 학습
minmax.fit(data_new)
```

```python
# transform : 데이터에 scaler 적용
minmax_scaled =  minmax.transform(data_new)
minmax_scaled
```

```python
# 보기 편하게 dataframe으로 변환
pd.DataFrame(minmax_scaled, columns= data_new.columns)
```

### 2) Standard Scaler

```python
standard = StandardScaler()
```

```python
# 종속 변수는 스케일링 시 제거해준다.
standard.fit(data_new.drop('Churn_Yes', axis=1))
```

```python
standard_scaled = standard.transform(data_new.drop('Churn_Yes', axis=1))
```

```python
pd.DataFrame(standard_scaled, columns = data_new.drop('Churn_Yes', axis=1).columns)
```

### 3) Robust Scaler

```python
rob = RobustScaler()
```

```python
rob.fit(data_new.drop('Churn_Yes', axis=1))
```

```python
robust_scaled = rob.transform(data_new.drop('Churn_Yes', axis=1))
```

```python
pd.DataFrame(robust_scaled, columns = data_new.drop('Churn_Yes', axis=1).columns)
```

### 최종적으로 Min-Max 스케일러로 데이터 저장

```python
scaled_data = pd.DataFrame(minmax_scaled, columns= data_new.columns)
scaled_data
```

# KNN 모델 만들기

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = scaled_data.drop('Churn_Yes', axis = 1)
y = scaled_data['Churn_Yes']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state =100)
```

```python
X_train
```

## Modeling

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
# knn 모델 학습
knn.fit(X_train, y_train)
```

```python
# test 데이터 예측
pred = knn.predict(X_test)
```

```python
pred[0:100]
```

```python
# 예측값과 실제값 비교
pd.DataFrame({'acutal_value' : y_test, 'pred_value' : pred }).head(30)
```

# 모델 정확도 확인 및 최적의 k값 찾기

## 모델 정확도 확인

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test, pred)
```

```python
confusion_matrix(y_test, pred)
```

## 최적의 K 값 구하기

```python
list(range(1, 101))[53]
```

```python
# for문으로 k값을 1~100까지 바꿔가면서 모델 예측
accuracy_list = []
for i in range(1, 101):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    accuracy_list.append(accuracy_score(y_test, pred))
```

```python
# 그래프 그려보기
plt.figure(figsize=(20,10))
sns.lineplot(x = range(1,101), y = accuracy_list, marker = 'o', markersize=10, markerfacecolor='red')

plt.xlabel('k', fontsize=20)
plt.ylabel('accuracy', fontsize=20)
```

```python
# accuracy 최대값
max(accuracy_list)
```

```python
# accuracy 최대값일 때의 k값
accuracy_list.index(max(accuracy_list))
```

```python
# 인덱스는 0부터 시작하므로, 최적의 k값은 index+1 = 54
```

```python
# argmax 함수로 accuracy 최대값 찾기
np.array(accuracy_list).argmax()
```

```python
accuracy_list[53]
```
